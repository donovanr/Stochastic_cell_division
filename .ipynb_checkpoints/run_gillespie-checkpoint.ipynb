## My Simulation

```julia
here = pwd()
module_dir = "$here/gillespiemodule"
push!(LOAD_PATH, module_dir);

using Distributions
using Gadfly; set_default_plot_size(6inch, 4inch);
using Plotly

using gillespie # library i wrote for gillespie ssa / cell division simulations
```

```julia
t_begin = 0;
t_end = 1*10^6;
dt = 1.0;
n_steps = iround((t_end - t_begin)/dt);

growth_rate = 0.0075;                       # protein production rate, (/prot/sec)

length_init = 50;                           # (nm) must be ints for this to run fast
prot_init = 50;                             # (number)
init_state = [length_init prot_init];       # x = (length, prot_count) -- length needs to be the 0th element

mu_l_threshold = 2.0*length_init;           # mean length at cell division (nm)
sigma_l_threshold = 0.25*mu_l_threshold;    # stdv of length at cell division (nm)

Stoich_mat = [1 0; 0 1];  

# cell regulation factor
regfac = 2.0;

function reaction_propensities(x,g)
    p = [g*x[1],g*x[2]]
    return p
end;
```

```julia
@time data = gillespie_even_spacing(n_steps, dt, init_state, Stoich_mat, reaction_propensities, growth_rate, mu_l_threshold, sigma_l_threshold, regfac);
```

## Plotting with Gadfly

This plotting library is somewhat slow the first time it's used

```julia
myplot = plot(x=data[1:10^4,1], y=data[1:10^4,3], Geom.line)
```

```julia
draw(PNG("myplot.png", 8inch, 6inch), myplot)
```

```julia
plot(x=data[:,3], Geom.histogram(bincount=iround( (maximum(data[:,3])-minimum(data[:,3]))/10 )))
```

## Plotting with Plotly

you need to change username to your username, and key to your login key

```julia
Plotly.signin("username", "key");
```

```julia
x = data[:,3]
plotdata = [
    [
    "name" => "Run 1", 
    "x" => x,
    "type" => "histogram",
        "histnorm" => "probability density", 
    "nbinsx" => 250
    ]
]

layout = [
    "title" => "Protein Histogram",
    "showlegend" => "False",
    "xaxis" => [
        "title" => "Protein Count",
        "autorange" => true
    ], 
    "yaxis" => [
        "title" => "Probability Density",
        "type" => "log",
        "autorange" => true
  ] 
]

response = Plotly.plot([plotdata], ["layout" => layout, "filename" => "basic-histogram", "fileopt" => "overwrite"])
plot_url = response["url"];
```

```julia
run(`open $plot_url`)
```
