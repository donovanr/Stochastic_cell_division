## Rate Feedback Regulation

```julia
t_begin = 0;
t_end = 1*10^5;
dt = 1.0;
n_steps = iround((t_end - t_begin)/dt);

growth_rate = 0.0075;                       # protein production rate, (/prot/sec)
reaction_rates = [growth_rate growth_rate]  # seperate rates for each reaction

length_init = 50;                           # (nm) must be ints for this to run fast
prot_init = 50;                             # (number)
init_state = [length_init prot_init];       # x = (length, prot_count) -- length needs to be the 0th element

mu_l_threshold = 2.0*length_init;           # mean length at cell division (nm)
sigma_l_threshold = 0.10*mu_l_threshold;    # stdv of length at cell division (nm)

Stoich_mat = [1 0; 0 1];  

# cell regulation factor
regfac = 0.3; # needs to be 0 < reggfac < 1
thresh_factor = 2.0

function reaction_propensities(x,r)
    p = [r[1]*x[1],r[2]*x[2]]
    return p
end;

#returns data as [t_list,z_list,r_list] as a row at each timepoint
```

```julia
function ijtok(i,j,imax,jmax)
    k = (i-1)*jmax + j
    return k
end

function ktoij(k,imax,jmax)
    j = (k-1)%jmax + 1
    i = int((k-j)/jmax + 1) 
    return [i,j]
end

# what range in parameter space are we scanning?
regfacs = [0.05,0.1,0.15]
thresh_factors = [2.0,2.5,3.0]

imax = length(regfacs)
jmax = length(thresh_factors)
kmax = imax*jmax


alldata = zeros(n_steps,length(init_state)+1,imax*jmax)


for i=1:imax
    for j=1:jmax
        #k = (i-1)*jmax + j # convert from 2-d indices to 1-d
        k = ijtok(i,j,imax,jmax)
        
        # run simulation
        data = gillespie_even_spacing(
            n_steps,
            dt,
            init_state,
            Stoich_mat,
            reaction_propensities,
            reaction_rates,
            mu_l_threshold,
            sigma_l_threshold,
            regfacs[i],
            thresh_factors[j]
        );
        alldata[:,:,k] = data[:,1:length(init_state)+1] # don't save reaction rates
                
    end
end
```

```julia
allhists = Array(Any, kmax)
alledges = Array(Any, kmax)

for k = 1:kmax
    v = vec(alldata[:,3,k])
    n = 100
    mn = 0
    mx = ceil(maximum(v)*1.05)
    rng = mx - mn;
    dx = ceil(rng/n)
    myedges = mn:dx:mx
    
    e,h = whist(v, myedges; density=true)
    alledges[k] = [e]
    allhists[k] = h
end
```

### Plotting with Plotly

```julia
traces = Array(Any, kmax)
# use col 3 for prot hists
for k=1:kmax
    idx = ktoij(k,imax,jmax)
        traces[k] = [
            "name" =>  string("regfac = ", idx[1], ", thresh = ", idx[2]), 
            "x" => alledges[k], 
            "y" => allhists[k], 
            "type" => "bar",
            "xaxis" => string("x", k), 
            "yaxis" => string("y", k),
            "showlegend" => false
        ]
end

axis_range = [1 => [0.0,0.3], 2 => [0.35,0.65], 3 => [0.7,1.0]]

layout = Dict()
for n=1:2*kmax
    if n%2 == 1
        k = int((n+1)/2)
        idx = ktoij(k,imax,jmax); i = idx[1]; j = idx[2]
        if j == 1
            layout[string("xaxis", k)] = [
                "domain" => axis_range[i],
                "anchor" => string("y", k),
                "autorange" => true,
                "title" => string("regfac = ", regfacs[i])
                ]
        else
            layout[string("xaxis", k)] = [
                "domain" => axis_range[i],
                "anchor" => string("y", k),
                "autorange" => true,
                ]
        end
    else
        k = int(n/2)
        idx = ktoij(k,imax,jmax); i = idx[1]; j = idx[2]
        if i == 1
            layout[string("yaxis", k)] = [
                "domain" => axis_range[j],
                "anchor" => string("x", k),
                "type" => "log",
                "autorange" => true,
                "title" => string("thresh_factor = ", thresh_factors[j])
                ]
        else
            layout[string("yaxis", k)] = [
                "domain" => axis_range[j],
                "anchor" => string("x", k),
                "type" => "log",
                "autorange" => true
                ]
        end
    end
end

layout["title"] = "Feedback Regulation: Protein Histogram"


Plotly.signin("donovanr", "jzib7flemh")

response = Plotly.plot([traces], ["layout" => layout, "filename" => "single cell histogram with feedback", "fileopt" => "overwrite"])
plot_url = response["url"]
#run(`open -a safari $plot_url`)
```

## Dumb Regulation

```julia
t_begin = 0;
t_end = 1*10^6;
dt = 1.0;
n_steps = iround((t_end - t_begin)/dt);

growth_rate = 0.0075;                       # protein production rate, (/prot/sec)
reaction_rates = [growth_rate growth_rate]  # seperate rates for each reaction
ratenoisefactor = 0.001

length_init = 50;                           # (nm) must be ints for this to run fast
prot_init = 50;                             # (number)
init_state = [length_init prot_init];       # x = (length, prot_count) -- length needs to be the 0th element

mu_l_threshold = 2.0*length_init;           # mean length at cell division (nm)
#sigma_l_threshold = 0.10*mu_l_threshold;    # stdv of length at cell division (nm)

Stoich_mat = [1 0; 0 1];

# cell regulation factor
#single_regfac = 0.5; 

function reaction_propensities(x,r)
    p = [r[1]*x[1],r[2]*x[2]]
    return p
end;

#returns data as [t_list,z_list,r_list] as a row at each timepoint
```

```julia
function ijtok(i,j,imax,jmax)
    k = (i-1)*jmax + j
    return k
end

function ktoij(k,imax,jmax)
    j = (k-1)%jmax + 1
    i = int((k-j)/jmax + 1) 
    return [i,j]
end

# what range in parameter space are we scanning?
single_regfacs = [0.5,1.0,2.0]
sigma_l_thresholds = [0.05,0.10,0.2]*mu_l_threshold

imax = length(single_regfacs)
jmax = length(single_regfacs)
kmax = imax*jmax


alldata = zeros(n_steps,length(init_state)+1,imax*jmax)


for i=1:imax
    for j=1:jmax
        #k = (i-1)*jmax + j # convert from 2-d indices to 1-d
        k = ijtok(i,j,imax,jmax)
        
        # run simulation
        data = gillespie_even_spacing_dumb(
            n_steps,
            dt,
            init_state,
            Stoich_mat,
            reaction_propensities,
            reaction_rates,
            mu_l_threshold,
            sigma_l_thresholds[j],
            single_regfacs[i],
            ratenoisefactor
        );
        alldata[:,:,k] = data[:,1:length(init_state)+1] # don't save reaction rates
                
    end
end
```

```julia
allhists = Array(Any, kmax)
alledges = Array(Any, kmax)

for k = 1:kmax
    v = vec(alldata[:,3,k])
    n = 100
    mn = 0
    mx = ceil(maximum(v)*1.05)
    rng = mx - mn;
    dx = ceil(rng/n)
    myedges = mn:dx:mx
    
    e,h = whist(v, myedges; density=true)
    alledges[k] = [e]
    allhists[k] = h
end

t_n = 10^4
short_time = zeros(t_n,kmax)
short_time_prot = zeros(t_n,kmax)
for k = 1:kmax
    short_time_prot[:,k] =  vec(alldata[1:t_n,3,k])
    short_time[:,k] =  vec(alldata[1:t_n,1,k])
end
```
